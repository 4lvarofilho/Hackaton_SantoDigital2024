Concatenação de todas as tabelas de vendas

```python
sales_df = pd.concat([sales_2015_df, sales_2016_df, sales_2017_df])
```

<h3>Parte 2: Pré-processamento de Dados</h3>

Detecção de valores faltantes

```python
missing_values = sales_df.isnull().sum()
print("Valores faltantes em sales_df:", missing_values)
```

Codificação one-hot para a coluna 'TerritoryKey'

```python
sales_df = pd.get_dummies(sales_df, columns=['TerritoryKey'])
```

Merge de dados entre a tabela Sales e Products

```python
merge_data = pd.merge(sales_df, products_df, on='ProductKey')
```

Codificação alvo para a coluna 'ProductKey'

```python
mean_sales = sales_df.groupby('ProductKey')['OrderQuantity'].mean()
sales_df['ProductKey_Target'] = sales_df['ProductKey'].map(mean_sales)
```

Padronização

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
sales_df[['OrderQuantity']] = scaler.fit_transform(sales_df[['OrderQuantity']])
products_df[['ProductPrice']] = scaler.fit_transform(products_df[['ProductPrice']])
```

```python
print(sales_df.columns)
```

<h3>Parte 3: Feature Engineering</h3>

Média móvel das vendas dos últimos 3 meses

```python
sales_df['Order_Moving_Avg'] = sales_df['OrderQuantity'].rolling(window=3).mean()
```

Variação percentual das vendas em relação ao período anterior

```python
sales_df['Sales_Pct_Change'] = sales_df['OrderQuantity'].pct_change()
```

Ajuste sazonal

```python
sales_df['Month'] = pd.to_datetime(sales_df['OrderDate']).dt.month
sales_df['Seasonal_Sales'] = sales_df['OrderQuantity'] / sales_df.groupby('Month')['OrderQuantity'].transform('mean')
```

Exemplo de incorporação de dados meteorológicos

```python
weather_data = pd.read_csv('/content/weather_data.csv')
sales_df = sales_df.merge(weather_data, on='OrderDate')
```

Erro por falta de samples (mas é um exemplo de implementação de transformações polinomiais)

```python
from sklearn.preprocessing import PolynomialFeatures

combined_df = sales_df.merge(products_df, on='ProductKey')

poly = PolynomialFeatures(degree=2, interaction_only=True)
interaction_features = poly.fit_transform(combined_df[['OrderQuantity', 'ProductPrice']])
df_interactions = pd.DataFrame(interaction_features, columns=poly.get_feature_names(['OrderQuantity', 'ProductPrice']))
combined_df = pd.concat([combined_df, df_interactions], axis=1)
```
